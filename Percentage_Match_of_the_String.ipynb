### Preprocessing

```python
# Preprocessing  

def remove_single_char_func(text, threshold=2):
  threshold = threshold
  words = word_tokenize(text) 
  texts = ' '.join([word for word in words if len(word)>threshold])
  return texts 



# remove_custom_stopwords:
custom_stops = ['a','able','about','across','after','all','almost','also','am','among','an','any',
                  'are','as','at','be','because','been','but','by','could','dear','did','do',
                  'does','either','else','ever','every','for','from','get','got','had','has','have','he','her',
                  'hers','him','his','how','however','i','if','into','it','its','just','let','like',
                  'likely','may','me','might','most','must','my','neither','no','nor','often',
                  'only','or','our','own','rather','said','say','says','she','should','since','so','some',
                  'than','that','the','their','them','then','there','these','they','this','tis','to','too','twas',
                  'us','wants','was','we','were','what','when','where','which','while','who','whom','why','will','with','would','yet','you','your'] 


def preprocessing(data, c): 

  # loewr case (both title and causes)
  data[c] = data[c].str.lower() 
  #remove multiple spaces 
  # data[c] = data[c].apply(lambda x: " ".join(x.split()))
  # remove contractions
  data[c] = data[c].apply(lambda x:contractions.fix(str(x)))
  # remove punctuation
  data[c] = data[c].apply(lambda x: re.sub('\W+\s*', ' ', x)) 
  # remove selected stopwords
  data[c] = data[c].apply(lambda x: " ".join([i for i in x.split() if i not in custom_stops]))
  # remove multiple spaces                   
  data[c] = data[c].apply(lambda x: " ".join(x.split()))
a = 'Public Title of Study'
b = 'List of Causes' 
preprocessing(df, a) 
preprocessing(df, b)
```

```python
# experimentation
x = fuzz.ratio('study on complications of diabetes and risk of heart attack', 'Diabetes and kidney diseases')
z= fuzz.partial_ratio('study on complications of diabetes and risk of heart attack', 'Diabetes and kidney diseases')
a= fuzz.token_set_ratio('study on complications of diabetes and risk of heart attack', 'Diabetes and kidney diseases')

b= fuzz.token_set_ratio('study on complications of diabetes and risk of heart attack', 'Diabetes mellitus')
c= fuzz.partial_ratio('study on complications of diabetes and risk of heart attack', 'Diabetes mellitus')
d= fuzz.ratio('study on complications of diabetes and risk of heart attack', 'Diabetes mellitus')
print('-----------------------------------------------------------------------')
e = fuzz.partial_ratio('study on complications of diabetes and risk of heart attack', 'Hypertensive heart disease')
f = fuzz.token_set_ratio('study on complications of diabetes and risk of heart attack', 'Hypertensive heart disease')

print('Ratio, partial_ratio, token_set_ratio-----------------------------------------------------------------------')
print('Diabetes and kidney diseases:')
print(x,z,a)
print('-----------------------------------------------------------------------')
print('token_set_ratio, partial_ratio, ratio :Diabetes mellitus ')
print(b,c,d)
print('-----------------------------------------------------------------------')
print('partial_ratio,token_set_ratio:Hypertensive heart disease')
print(e, f) 
```

* Lets consider partial_ratio for this dataset, which is neither fully sensitive like token_set_ratio nor too specific like ratio.  

```python
# Get the titles 
df_res = pd.DataFrame() 
df_res['Public Title of Study'] = df['Public Title of Study']
# Remove the Null Value observed. 
df_res = df_res.drop([2,151])
# Get the causes
causes = df['List of Causes'].unique() 
# remove the last nan value 
causes = causes[:255]  
```

```python
def fuzz_percentage(x,cause): 
  e = fuzz.partial_ratio(x,cause)
  return e

for i in range(len(causes)): 
  df_res['Cause'+ str(i)] = causes[i] 
  df_res['Cause'+ str(i)+ ' % Match'] = df['Public Title of Study'].apply(lambda x: fuzz_percentage(x,causes[i]))
```

### Ratio

```python
# Get the titles 
df_ratio = pd.DataFrame() 
df_ratio['Public Title of Study'] = df['Public Title of Study']
# Remove the Null Value observed. 
df_ratio = df_ratio.drop([2,151])
# Get the causes
causes = df['List of Causes'].unique() 
# remove the last nan value 
causes = causes[:255]  



def fuzz_percentage(x,cause): 
  e = fuzz.ratio(x,cause)
  return e

for i in range(len(causes)): 
  df_ratio['Cause'+ str(i)] = causes[i] 
  df_ratio['Cause'+ str(i)+ ' % Match'] = df['Public Title of Study'].apply(lambda x: fuzz_percentage(x,causes[i]))
```

### Token Sort Ratio

```python
# Get the titles 
df_token_ratio = pd.DataFrame() 
df_token_ratio['Public Title of Study'] = df['Public Title of Study']
# Remove the Null Value observed. 
df_token_ratio = df_token_ratio.drop([2,151])
# Get the causes
causes = df['List of Causes'].unique() 
# remove the last nan value 
causes = causes[:255]  



def fuzz_percentage(x,cause): 
  e = fuzz.token_sort_ratio(x,cause)
  return e

for i in range(len(causes)): 
  df_token_ratio['Cause'+ str(i)] = causes[i] 
  df_token_ratio['Cause'+ str(i)+ ' % Match'] = df['Public Title of Study'].apply(lambda x: fuzz_percentage(x,causes[i])) 

df_token_ratio.to_csv(folder_path + 'token_ratio_result.csv')

kc_val_tr = df_token_ratio['Cause122 % Match'].max()
print('% of kidney cancer:')
print(kc_val_tr)
dk_val_tr = df_token_ratio['Cause57 % Match'].max()
hd_val_tr = df_token_ratio['Cause109 % Match'].max()
print('% of diabetes and kidney disease:')
print(dk_val_tr)
print('% of hypertensive_heart_disease:')
print(hd_val_tr)
kidney_cancer_matches_tr = df_token_ratio[df_token_ratio['Cause122 % Match']== kc_val_tr]['Public Title of Study']
diabetes_kidney_matches_tr = df_token_ratio[df_token_ratio['Cause57 % Match']== dk_val_tr]['Public Title of Study']
hypertensive_heart_matches_tr = df_token_ratio[df_token_ratio['Cause109 % Match']== hd_val_tr]['Public Title of Study']

print('Kidney cancer:')
for i in kidney_cancer_matches_tr:
  print(i) 
print('------------------------------------------------')
print('Diabetes and kidney disease:')
for i in diabetes_kidney_matches_tr:
  print(i) 
print('------------------------------------------------')
print('hypertensive_heart_disease:')
for i in hypertensive_heart_matches_tr:
  print(i) 
```
